```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the entire dataset

```python
data = pd.read_csv("D:\\Python\\matplotlib\\Rainfall\\ne-India_rainfall_act_dep_1901_2016.csv", index_col=0)
```

# Visualization of the entire dataset

```python
data.plot()
plt.legend(loc="center", bbox_to_anchor=(0.5, 1.4))
plt.ylabel('rainfall in mm')
plt.show()
```

# Visualization of the recent 10 years (2007-2016)

```python
df=data.tail(10)
df.plot()
plt.ylabel('rainfall in mm')
plt.legend(loc="center", bbox_to_anchor=(0.5, 1.4))
plt.show()
```

# Visualization according to user

```python
yStart=int(input('From year: '))
yEnd=int(input('To year: '))
df = data.loc[yStart:yEnd, ['Actual Rainfall: JUN', 'Actual Rainfall: JUL', 'Actual Rainfall: AUG', 'Actual Rainfall: SEPT', 'Actual Rainfall: JUN-SEPT']]
df.plot()
plt.show()
```

# Actual Rainfall vs Departure Percentage (JUN-SEPT)

```python
sns.relplot(x="Actual Rainfall: JUN-SEPT", y="Departure Percentage: JUN-SEPT", data=data)
```
